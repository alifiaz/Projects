# APRIORI ALGORITHM
#### Instead of using Package or libraries, we will use aprior.py file

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('Market_Basket_Optimisation.csv', header=None) #No column header
dataset.head(4)
```

```python
dataset.shape
```

```python
transactions=[] #Empty List
# The dataset should be in the form list of list
```

```python
for i in range(0,7501):#Loop through all transactions, 0 to 7500
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

## Training Apriori on the dataset

```python
from apyori import apriori
```

```python
rules= apriori(transactions,min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)

#rules will be the output of apriori function

#min_confidence=0.2, thus all rules we obtain will be true atleast 20% of the time

# min_support: support of a set of items 'i' is equal to no of transactions containing items 'i'
#divided by the total no of transactions, We consider products purchased 03times a day, 
#07 times a week, 3*7/7500=0.003

# min_lenght is 02 as we want a minimum of two products transaction in basket
```

## Visualising the Results

```python
results=list(rules)
```
